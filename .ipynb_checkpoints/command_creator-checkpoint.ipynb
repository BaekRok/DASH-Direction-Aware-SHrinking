```python
def command_merge(commands):
    com = ""
    for i, command in enumerate(commands):
        if i + 1 == len(commands):
            com += command
        else:
            com += command + ' && '
    print(com)
```

```python
def get_command(nohup=False):
    if nohup:
        commands = ['conda activate torch && nohup', ' python ','Online_Learning/main.py ',
                ' --full_train=',
                str(full_train),
                ' --max_epoch=',
                str(max_epoch),
                ' --depth=',
                str(depth),
                ' --width=',
                str(width),
                ' --epoch=',
                str(epoch),
                ' --model=', 
                model,
                ' --dataset=', 
                dataset,
                ' --batch_size=', 
                str(batch_size), 
                ' --optimizer_type=',
                optimizer_type,
                ' --learning_rate=', 
                str(learning_rate), 
                ' --momentum=',
                str(momentum), 
                ' --gpu=', 
                str(gpu), 
                ' --sp_lambda=',
                str(sp_lambda),
                ' --last_sshrink=',
                str(last_sshrink),
                ' --es_threshold=', 
                str(es_threshold),
                ' --es_temperature=',
                str(es_temperature),
                ' --es_grad_alpha=',
                str(es_grad_alpha),
                ' --alpha=',
                str(alpha),
                ' --task=',
                str(task),
                ' --converge_acc=',
                str(converge_acc),
                ' --l2_init_lambda=',
                str(l2_init_lambda),
                ' --test_batch_size=',
                str(test_batch_size),
                ' --rho=',
                str(rho),
                ' --scheduler_tmax=',
                str(scheduler_tmax),
                ' --weight_decay=',
                str(weight_decay),
                ' --train_type=',
                str(train_type),
                ' --seed=', 
                str(seed), 
                ' &'] 
    else:
        commands = ['conda activate torch &&', ' python ','Online_Learning/main.py ',
                    ' --full_train=',
                    str(full_train),
                    ' --max_epoch=',
                    str(max_epoch),
                    ' --depth=',
                    str(depth),
                    ' --width=',
                    str(width),
                    ' --epoch=',
                    str(epoch),
                    ' --model=', 
                    model, 
                    ' --dataset=', 
                    dataset, 
                    ' --batch_size=', 
                    str(batch_size), 
                    ' --optimizer_type=',
                    optimizer_type,
                    ' --learning_rate=', 
                    str(learning_rate), 
                    ' --momentum=',
                    str(momentum), 
                    ' --gpu=', 
                    str(gpu), 
                    ' --sp_lambda=',
                    str(sp_lambda),
                    ' --last_sshrink=',
                    str(last_sshrink),
                    ' --es_threshold=', 
                    str(es_threshold),
                    ' --es_temperature=',
                    str(es_temperature),
                    ' --es_grad_alpha=',
                    str(es_grad_alpha),
                    ' --alpha=',
                    str(alpha),
                    ' --task=',
                    str(task),
                    ' --converge_acc=',
                    str(converge_acc),
                    ' --l2_init_lambda=',
                    str(l2_init_lambda),
                    ' --test_batch_size=',
                    str(test_batch_size),
                    ' --rho=',
                    str(rho),
                    ' --scheduler_tmax=',
                    str(scheduler_tmax),
                    ' --weight_decay=',
                    str(weight_decay),
                    ' --train_type=',
                    str(train_type),
                    ' --seed=', 
                    str(seed), 
                    ''] 
    return commands
```
